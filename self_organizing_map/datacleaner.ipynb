```python
import numpy as np
#import matplotlib.pyplot as plt
import pandas as pd
import glob
```

```python
#food = pd.read_csv("csvfiles/samplefoods.csv",header=None)
food = pd.read_csv("csvfiles/masktest.csv",header=None)
foodlist = food.T.values.tolist()[0]
print(foodlist)
mask = r'|'.join(foodlist)

mac_path = r'csvfiles/macronutrients'
min_path = r'csvfiles/minerals'
vit_path = r'csvfiles/vitamins'
def import_csv(local_path):
    filenames = glob.glob(local_path + "/*.csv")
    dfs = [pd.read_csv(filename,encoding="latin",skiprows=1, na_values='0',low_memory=False) for filename in filenames]
    big_frame = pd.concat(dfs, ignore_index=True,axis=0)
    return big_frame
```

```python
macros = import_csv(mac_path)
minerals = import_csv(min_path)
vitamins = import_csv(vit_path)

#macros
```

```python
macros.to_csv("csvfiles/all/allmacros.csv")
minerals.to_csv("csvfiles/all/allminerals.csv")
vitamins.to_csv("csvfiles/all/allvitamins.csv")
```

```python
macros = macros[macros['Description'].str.contains(mask, na = True)]
minerals = minerals[minerals['Description'].str.contains(mask, na = True)]
vitamins = vitamins[vitamins['Description'].str.contains(mask, na = True)]
macros
```

```python
def all_import_csv(local_path):
    filenames = glob.glob(local_path + "/*.csv")
    dfs = [pd.read_csv(filename,encoding="latin", na_values='0',low_memory=False) for filename in filenames]
    big_frame = pd.concat(dfs, ignore_index=True, axis=0)
    return big_frame
```

```python
concat_path = r'csvfiles/concat'
food_table = all_import_csv(concat_path)
```

```python
food_table = food_table.drop(columns=['Unnamed: 0'])
food_table.to_csv("csvfiles/concat.csv")
food_table
```
