```python
import matplotlib.pyplot as plt
import numpy as np
import numba as nb
from numba import jit,njit
```

```python
xfixed=[]
tarray=[]
def chaos_map(x,r):
    for t in np.arange(0,20):
        x=r*x*(1-x)
        xfixed.append(x)
        tarray.append(t)
    return xfixed, tarray
```

```python
chaos_map(0.5,3.5)
plt.plot(tarray,xfixed)
plt.title("Chaos Map",fontsize=20)
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel("t", fontsize=20)
plt.ylabel("x(t)", fontsize=20)
```

```python
rarray = []
xarray = []
def logistic(x):
    for r in np.arange(2.8,4,.001):
        for t in nb.prange(200):
            x=r*x*(1-x)
            rarray.append(r)
            xarray.append(x)
```

```python
logistic(0.5)
plt.title("Bifrucation Diagram of Logistic Map",fontsize=20)
plt.scatter(rarray,xarray,marker='o',s=0.001,alpha=1)
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel("r", fontsize=20)
plt.ylabel("x", fontsize=20)
```
