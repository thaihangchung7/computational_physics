```python
import matplotlib.pyplot as plt
import numpy as np
import numba as nb
from numba import jit,njit, prange
import matplotlib.animation as ani
from matplotlib.animation import FuncAnimation
```

```python
xfixed=[]
tarray=[]
def chaos_map(x,r):
    for t in np.arange(0,20):
        x=r*x*(1-x)
        xfixed.append(x)
        tarray.append(t)
    return xfixed, tarray
```

```python
chaos_map(0.5,3.5)
plt.plot(tarray,xfixed)
plt.title("Chaos Map",fontsize=20)
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel("t", fontsize=20)
plt.ylabel("x(t)", fontsize=20)
```

```python
rarray = []
xarray = []

def logistic(x):
    for r in np.arange(2.8,4,.001):
        for t in nb.prange(2000):
            x=r*x*(1-x)
            rarray.append(r)
            xarray.append(x)
```

```python
logistic(0.5)
plt.title("Bifrucation Diagram of Logistic Map",fontsize=20)
plt.scatter(rarray,xarray,marker='o',s=0.001,alpha=1)
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel("r", fontsize=20)
plt.ylabel("x", fontsize=20)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
scatter, = ax.plot(rarray,xarray)

def animate_func(i):
    scatter.set_data(rarray[i],xarray[i])
    return scatter,

#ani = FuncAnimation(fig, animate_func, interval=100, frames=N)
ani = ani.FuncAnimation(fig, animate_func, len(rarray), interval = 100, blit=False)
ani.save('logi.gif', writer='ffmpeg')
plt.show()
```

```python
x_init = 0.5
N = 200
r = np.arange(2.4, 4., .001)

@jit(nopython=True)
def numba_logistic(x0,r):
    x = np.full_like(r,x0)
    print(x)
    for n in range(30000):
        x = x * r * (1-x)
        
    final_list = np.empty((N + 1, r.shape[0]), dtype=np.float64)
    final_list[0,:] = x
    for n in range(1,N + 1):
        final_list[n,:] = final_list[n-1] * r * (1 - final_list[n-1])
    out = np.empty_like(final_list)
    for n in range(r.shape[0]):
        out[:,n] = np.sort(final_list[:,n])
        
    return out
```
